## inisialisasi tabel Q dengan 0

```python
listx = np.zeros(100)
Q = {'N': listx, 'E': listx, 'S': listx, 'W': listx}
tabelQ = pd.DataFrame(data=Q)
```

## inisialisasi parameter yang digunakan yaitu gamma, jumlah episodes dan alpha

```python
gamma= 0.7
maxEpisodes= 10000
alpha = 0.25
```

## fungsi untuk mengambil action secara acak dengan memperhatikan langkah yang memungkinkan 

```python
def randomAction(state1, state2):
    act = ''
    finish = False

    while finish == False:
        stateBaru1 = state1
        stateBaru2 = state2
        action = random.randint(1, 4)
        # print('random: ', action)
        if (action == 1): #Atas
            act = 'N'
            stateBaru2 = state2 -1
            if (stateBaru2  <= 9 and stateBaru2 >= 0):
                finish = True
        elif (action == 2): #Kanan
            stateBaru1 = state1 + 1
            act = 'E'
            if (stateBaru1  <= 9 and stateBaru1 >= 0):
                finish = True
        elif (action == 3): #Kiri
            act = 'W'
            stateBaru1 = state1 -1
            if (stateBaru1  <= 9 and stateBaru1 >= 0):
                finish = True
        elif (action == 4): #Bawah
            act ='S'
            stateBaru2 = state2 + 1
            if (stateBaru2 <= 9 and stateBaru2 >= 0):
                finish = True
    result = [stateBaru1, stateBaru2, act]
    return result
```

## Fungsi untuk mengeksplorasi dengan mengupdate tabel Q dengan rumus yang ada sampai jumlah episode yang telah didefinisikan tercapai

```python
listReward = []
for i in range(0, maxEpisodes):
    currentState1 = 0
    currentState2 = 9
    print('Episodenya: ', i)
    reward = 0
    while currentState1 != 9 or currentState2 != 0:
        action = randomAction(currentState1, currentState2)

        
        tabelQ[action[2]][int(str(currentState1) + str(currentState2))] = tabelQ[action[2]][int(str(currentState1) + str(currentState2))] + (alpha * (tabelR[action[0]][action[1]] + (gamma * max(tabelQ.iloc[int(str(action[0]) + str(action[1]))]) -  tabelQ[action[2]][int(str(currentState1) + str(currentState2))]))) 
        reward = reward + tabelR[action[0]][action[1]]
        currentState1 = action[0]
        currentState2 = action[1]
    print(currentState1 ,' , ', currentState2)
    print('Reward: ', reward)
    listReward.append(reward)
print('Reward Maximum: ', max(listReward))
```

# Proses Exploitas dimulai

### Fungsi untuk melakukan action yang diperlukan

```python
def action(state1, state2, act):
    if (act == 'W'):
        state1 = state1 - 1
        print('W')
    elif (act == 'E'):
        state1 = state1 + 1
        print('E')
    elif (act =='N'):
        print('N')
        state2 = state2 - 1
    elif (act == 'S'):
        print('S')
        state2 = state2 + 1
        
    return [state1, state2]
```

### Proses exploitasi sampai dengan mencapai kordinat 9,0 yang mana merupakan titik finish dengan reward 100

```python
currentState1 = 0
currentState2 = 9

tabelQ = tabelQ.replace(0, -99)

while currentState1 != 9 or currentState2 != 0:
    # Mengambil nilai maksimum dari tabel Q
    maxReward = max(tabelQ.iloc[int(str(currentState1) + str(currentState2))])
        
    if tabelQ['N'][int(str(currentState1) + str(currentState2))] == maxReward:
        newState = action(currentState1, currentState2, 'N')
    elif tabelQ['S'][int(str(currentState1) + str(currentState2))] == maxReward:
        newState = action(currentState1, currentState2, 'S')
    elif tabelQ['E'][int(str(currentState1) + str(currentState2))] == maxReward:
        newState = action(currentState1, currentState2, 'E')
    elif tabelQ['W'][int(str(currentState1) + str(currentState2))] == maxReward:
        newState = action(currentState1, currentState2, 'W')
    # Update posisi
    currentState1 = newState[0]
    currentState2 = newState[1]
    print('State 1: ', currentState1)
    print('State 2: ', currentState2)
```
